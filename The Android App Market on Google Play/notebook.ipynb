```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
#Import the CSV into pandas
apps = pd.read_csv('datasets/apps.csv')
#Clean the dataset
apps.Installs = apps.Installs.str.replace(',','').str[:-1]
apps.Installs = apps.Installs.replace('','0')
apps.Installs = apps.Installs.astype(int)
#Print DF stats
apps.info()
print("\n")
apps.head()
```

```python
#Find the number of apps in each Category, the average price, and the average rating
app_category_info = apps.groupby('Category').agg({'App':'count', 'Price':'mean', 'Rating':'mean'})
app_category_info.columns = ['Number of Apps','Average price', 'Average rating']
app_category_info
```

```python
#Filter the Finance Apps
finance_apps = apps[apps['Category']=='FINANCE']
finance_apps.head()
```

```python
#Import user reviews into pandas DataFrame
reviews = pd.read_csv('datasets/user_reviews.csv')
reviews.info()
reviews.head()
```

```python
#Calculate the Sentiment score for each app
grouped_reviews = reviews.groupby('App').agg({'Sentiment Score':'mean'})
grouped_reviews.head()
```

```python
#Find the sentiment score for each finance app
finance_feedback = finance_apps.merge(grouped_reviews, on='App')
sorted_finance_fb = finance_feedback.sort_values(by=['Sentiment Score'], ascending=False)
#Get the TOP 10 finance apps
top_10_user_feedback = sorted_finance_fb.head(10)[['App','Sentiment Score']]
top_10_user_feedback = top_10_user_feedback.set_index('App')
top_10_user_feedback
```
