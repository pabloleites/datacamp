```python
# import all modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

grant_file = 'datasets/schoolimprovement2010grants.csv'

# Read in the DataFrame
df = pd.read_csv(grant_file)
```

## Pandas Histogram

```python
# Display pandas histogram
df['Award_Amount'].plot.hist()
plt.show()
```

## Seaborn Distplot

```python
# Display a Seaborn distplot
sns.distplot(df['Award_Amount'])
plt.show()
```

## Histogram with Seaborn
The distplot() function will return a Kernel Density Estimate (KDE) by default. 
The KDE helps to smooth the distribution and is a useful way to look at the data. 
However, Seaborn can also support the more standard histogram approach if that is more meaningful for your analysis.

```python
# Create a distplot
sns.distplot(df['Award_Amount'],
             kde=False,
             bins=20)

# Display the plot
plt.show()
```

## Rug plot and kde shading

```python
# Create a distplot of the Award Amount
sns.distplot(df['Award_Amount'],
             hist=False,
             rug=True,
             kde_kws={'shade':True})

# Plot the results
plt.show()
```

## Regression Plot

```python
# Import dataframe
df = pd.read_csv('datasets/insurance_premiums.csv')

# Create a regression plot of premiums vs. insurance_losses
sns.regplot(x='insurance_losses', y='premiums', data=df)

# Display the plot
plt.show()
```

```python
# Create an lmplot of premiums vs. insurance_losses
sns.lmplot(data=df, y='premiums', x='insurance_losses')

# Display the second plot
plt.show()
```

```python
# Create a regression plot using hue
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           hue="Region")

# Show the results
plt.show()
```

```python
# Create a regression plot with multiple columns
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           col="Region")

# Show the plot
plt.show()
```
